```python
import pandas as pd
```

```python
df=pd.read_csv("Housing.csv")
df=df.dropna()
```

```python
df.head(2)
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
x=df.drop('price',axis=1)
```

```python
y=df['price']
```

```python
from numpy import size
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
L=LinearRegression()
l_model=L.fit(x_train,y_train)
```

```python
l_prediction=l_model.predict(x_test)
```

```python
l_prediction
```

```python
x_test.head()
```

```python
x_test=x_test.reset_index(drop=True)
```

```python
x_test[0:1]
```

```python
l_model.predict(x_test[0:1])
```

```python
y_test=y_test.reset_index(drop=True)
```

```python
y_test[0:1]
```

```python
y_test.head()
```

```python
from xgboost import XGBRegressor
```

```python
x=XGBRegressor()
x_model=x.fit(x_train,y_train)
```

```python
x_model.predict(x_test[0:1])
```

```python
x_pre=x_model.predict(x_test)
```

```python
y_test[0:1]
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

rmse_lr = np.sqrt(mean_squared_error(y_test, l_prediction))
r2_lr = r2_score(y_test, l_prediction)

rmse_xgb = np.sqrt(mean_squared_error(y_test, x_pre))
r2_xgb = r2_score(y_test, x_pre)

print("Linear Regression RMSE:", rmse_lr)
print("Linear Regression R2:", r2_lr)

print("XGBoost RMSE:", rmse_xgb)
print("XGBoost R2:", r2_xgb)
```
